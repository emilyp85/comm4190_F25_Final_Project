# Pitch — Campus Housing Troubleshooter

## Purpose 
Give students a fast, correct **triage path** for housing problems by turning a messy description (e.g., “mold, roommate won’t respond, I have asthma”) into a **procedural, grounded plan** with the right office, exact steps, scripts, and what to bring.

- Issue types: **maintenance**, **roommate conflict**, **noise/quiet hours**, **temporary housing**, **accessibility accommodations**.  

## The Specific Issue It Targets
Students don’t know which **pathway** applies to their situation (maintenance ticket vs. RA mediation vs. Accessibility Services vs. temporary housing). Combining this with the social stress of not knowing the Penn community as a first/second year or even a transfer student we see a clear demand for this product. Policy info is scattered across web-pages and as a result, they contact the wrong office, use the wrong language, or miss documentation—**delaying resolution and risking health/safety**.

## Who It’s For
- On-campus residents (undergrad/grad), RAs, and student affairs staff.  
- Especially helpful for **after-hours** or **multi-factor** issues (health + facilities + conduct).

## Inputs -> Outputs
- **Input:** Free-text problem description (can include urgency, time of day, disability considerations).  
- **Output:**  
  1) **Step-by-step plan** (who to contact, where/when, what to bring, how to submit)  
  2) **Email/message script** with the right keywords  
  3) **Timeline & fees** (if any)  
  4) **After-hours alternative**  
  5) **Confirmation checklist** (what to verify before going)

- Issue types: **maintenance**, **roommate conflict**, **noise/quiet hours**, **temporary housing**, **accessibility accommodations**.
