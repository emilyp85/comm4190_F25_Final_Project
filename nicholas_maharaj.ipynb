# Pitch

People spend a lot of time trying to find an outfit for the day or a specific event. So I want to create Outfit.AI, a website that uses a large LLM to help people curate outfits for specific occassions. Users can upload pictures of their clothes and the event they are going to (Sunday brunch, class, work, etc.) and the LLM will create an outfit for them. It will also provide links to clothes to buy online to fill gaps in their closet.
