### **GOALS**

1. **Understand the event context**

2. **Obtain and respect consent for image use**

4. **Convert analysis into style guidance**  
   Translate those observations into:
   - Suggested silhouettes (e.g., wrap, A-line, fit-and-flare)

   Pull dress options that fit:
   - User‚Äôs size range and budget

6. **Integrate existing wardrobe**  
   Suggest ways to style each dress with **shoes, bags, and accessories the user already owns** when possible.

7. **Reduce decision fatigue and build confidence**  
   Present a **shortlist of clearly explained options** and allow the user to favorite a few, making the decision easier.

8. **Update style preferences for the future (optional)**  
   Learn from what the user favorites or rejects to refine future suggestions (e.g., preferred necklines, lengths, silhouettes).

#### **STEP 3 ‚Äî IMAGE ANALYSIS (ABSTRACT, NON-JUDGMENTAL)**

Fashion.AI must **not**:

- Infer race, ethnicity, or other sensitive traits

#### **STEP 4 ‚Äî STYLE GUIDANCE FROM ANALYSIS**

4. Fashion.AI converts image analysis + user constraints into **style guidance**, for example:

- Suggest **silhouettes**:
  - ‚ÄúWrap dresses or A-line shapes that skim the body‚Äù  
  - ‚ÄúFit-and-flare styles that define the waist without feeling tight‚Äù

- Suggest **necklines**:

- Suggest **hemlines**:
  - Knee, midi, or full-length based on:

- Suggest **comfort-first choices**:

6. Fashion.AI presents a **shortlist of dresses** (e.g., 3‚Äì7 options), each with:

### **ERROR HANDLING & ADVERSARIAL INPUTS**

#### **Possible Complications or Errors**

- **Poor-quality photos**
  - Low lighting, heavy filters, blurry images, heavy occlusions (e.g., large coat, cluttered mirror)

- **Non-standard poses or garments**
  - Dramatic poses, sitting/kneeling, or clothing that hides most of the body (e.g., oversized outerwear, cultural garments where standard cuts aren‚Äôt visible)

- **Wrong subject / others in the image**
  - User uploads a photo of another person
  - Group photos or images that clearly are not of the user

- **Explicit or adversarial content**
  - Nudity, explicit content
  - Attempts to trick the model into unsafe commentary or analysis
  - Requests to infer sensitive traits: weight, health, race, attractiveness rankings

#### **How Fashion.AI Should Respond**

Fashion.AI should **fail safely**:

- **Low-confidence analysis**
  - If the selfie is low-quality or heavily occluded:
    - Explain that accurate style guidance based on the image is limited.
    - Rely more on text-based preferences and self-reported fit issues.
    - Example:

- **Images of others or unknown subjects**
  - Politely decline to analyze:
    - ‚ÄúI can only give style advice based on photos of you, not other people.‚Äù
  - Ask the user to confirm and re-upload a selfie of themselves if they wish.

- **Explicit or inappropriate content**
  - Refuse to analyze:
    - ‚ÄúI‚Äôm not able to analyze explicit images.‚Äù
  - Offer to help with **text-based** outfit advice instead.

- **Requests for sensitive inferences**
  - If the user asks about weight, body type ranking, or attractiveness:
    - Decline to judge or guess.
    - Recenter on:

      - Comfort
