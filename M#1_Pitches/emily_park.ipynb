# Pitch
Students spend so much time trying to understand and organize class notes, readings, and lectures — it can get overwhelming. My idea is to fix that.

I’m creating SmartStudy, an app that uses a large language model (LLM) — basically, advanced AI that understands and writes text — to help students study smarter. It can summarize long readings, pull out key ideas, and even make practice questions based on what you’re learning.

In short, SmartStudy saves time, reduces stress, and helps students actually learn instead of just cramming.
