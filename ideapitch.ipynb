# Idea Pitch

Fashion.AI is an LLM model designed to help users craft their wardrobe and outfits to fit their specific needs and events. The LLM will help users design outfits—from shoes to accessories—to stay on theme for birthday parties, look sharp for interviews, etc. The LLM will also aid users in purchase decisions to help supplement their wardrobe to create a well-rounded closet. Additionally, the LLM can help users find pieces of clothing that complement their body type and desired style or aesthetic. Fashion.AI is meant to help users who are uncomfortable and unconfident with styling themselves, consumers overwhelmed with the fashion industry, or fashionistas who want to get ahead of trends. 

# Description of Target Users

The target users of Fashion.AI are individuals who place importance on their personal appearance and seek to present themselves well in various settings. This includes people who are interested in fashion, those who want to improve their sense of style, or anyone looking for convenient outfit recommendations. Because Fashion.AI is capable of generating outfit combinations that are not limited by gender norms, it is accessible and inclusive to all users, regardless of gender identity or expression. By offering personalized styling suggestions, Fashion.AI appeals to a broad audience that values confidence, creativity, and self-expression through clothing. We expect users to be mostly women in their twenties or late teen years, however. This is due to the fact that women at that age tend to care more about their appearance and outfits than other demographics.

# More Detailed Summary of what the application will do:

Fashion.AI is designed to help users make smart fashion choices, whether that’s styling their next outfit, finding new pieces to add to their wardrobe, or providing new ways to accessorize. What users will do is upload pictures of their clothing, everything from pants and shirts to even hats and accessories if the user wishes. Once the LLM has access to the clothes the user wishes the upload, the LLM is ready to tackle any question the user may have. The user can provide descriptions of party themes (ex. Kelly’s 1920’s speakeasy 21st birthday or Michael’s garden brunch) and the LLM will curate an outfit to fit the desired context. The user can ask for Halloween costumes, everyday outfits, business professional, etc and the LLM will provide an output of their clothes to form an outfit. The user can also tell the LLM they want to shop at a specific store and the LLM will provide recommendations of products to purchase, either to supplement their wardrobe or align with their current style.

# What data (from users, web, other sources, etc.) will be needed: 

The LLM will be trained (and continually refined to keep up on current trends) on a mix of fashion-specific text and visual data. These will be pulled from the web in the form of advice from fashion designers in blogs, magazines, etc.  In addition, the LLM will be trained on real-world shopping + sizing information. That includes labeled outfit images with metadata (items, colors, silhouettes, occasions, body types), product catalogs from diverse brands (descriptions, fabrics, size charts, price points), user-style profiles and feedback (what was recommended, chosen, or rejected), and guidelines on dress codes and event-appropriate attire. The LLM will have access to popular brands size charts in order to accurately provide users sizes of clothing that will fit them best depending on the brand that they are shopping with (H&M, American eagle, etc etc). This data will be diverse across genders, cultures, sizes, budgets, and aesthetics to avoid biased suggestions. In addition, the user will provide the LLM with the items currently in their wardrobe, their budget, their body type, location (so that the LLM can look into that region's climate and daily weather forecasts), and styles/colors/patterns that the user enjoys. 

# How Our Product Will Differ

Unlike most existing styling platforms that separate users by gender or focus on a narrow set of trends, Fashion.AI takes an inclusive approach. It can generate outfits for anyone, regardless of gender identity, body type, or personal style. This flexibility allows users to explore a wide range of looks and express themselves freely without being confined by traditional fashion categories. Furthermore, while other apps rely on pre-set outfit templates, Fashion.AI uses advanced artificial intelligence to create fully customized looks based on user input—such as occasion, mood, or aesthetic preference. The result is a more personalized, intuitive styling experience that feels like working with a real stylist rather than browsing a catalog that isn’t made for you. Instead of promoting specific brands or requiring purchases, Fashion.AI will have the option to generate creative outfit ideas that users can recreate with what they already own. This makes Fashion.AI both affordable and empowering for everyday users.
